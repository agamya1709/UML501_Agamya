```python
def impact(X):
    if X>=80:

        return 0
    elif 60<=X<80:

        return 1
    elif 40<=X<60:

        return 2
    elif 20<=X<40:

        return 3
    else:

        return 4
```

```python

df['TargetNew']=df['HealthImpactScore'].apply(impact)
```

```python
from sklearn.model_selection import train_test_split
```

```python

X=df.drop(['HealthImpactScore', 'TargetNew'],axis=1)
y=df['TargetNew']
```

```python

y=np.array(y)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y.reshape(-1, 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python

X_train.shape
```

```python
y
```

```python
model = Sequential()
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(4, activation='softmax'))  # 4 because there are 4 classes
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x=X_train,
          y=y_train,
          epochs=30,
          validation_data=(X_test, y_test), verbose=2
          )
```

```python
loss=pd.DataFrame(model.history.history)
```

```python

plt.figure(figsize=(9,7))
sns.lineplot(loss)
```

```python

from sklearn.metrics import confusion_matrix,classification_report
```

```python
predictions=(model.predict(X_test)>0.5).astype('int32')
```

```python
print(classification_report(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
losses,accuracy=model.evaluate(X_test,y_test)
print(f"Test Loss Is: {round(losses,2)},\nTest Accuracy Is: {round(accuracy,2)*100:.0f}%")
```
